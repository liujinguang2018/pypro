```python
#散点图， 这种图表中的一个数据集作为其他数据集的x值
#用于绘制一个金融时间序列的收益和另外一个时间序列收益的对比
y = np.random.standard_normal((1000, 2))
plt.figure(figsize=(12, 6))
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title("scatter plot")
```

```python
#使用matplotlib中提供的特殊函数
plt.figure(figsize=(12, 6))
plt.scatter(y[:, 0], y[:, 1], marker='o')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('scatter plot')
```

```python
#加入第三维，通过不同颜色进行可视化，并采用彩条加以描述
c = np.random.randint(0, 10, len(y))
plt.figure(figsize=(12, 6))
plt.scatter(y[:, 0], y[:, 1], c=c, marker='o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('scatter plot')
```

```python
y = np.random.standard_normal((1000, 2))  #正态分布
#直方图常用于显示金融收益
plt.figure(figsize=(12, 6))
#bins: 数据组数
plt.hist(y, label=['1st', '2nd'], bins=20)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
#数据堆叠
plt.figure(figsize=(12, 6))
#bins: 数据组数
plt.hist(y, label=['1st', '2nd'], bins=20, stacked=True, rwidth=0.8, color=['b', 'g'])
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
#箱型图
fig, ax = plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax, xticklabels=['1st', '2nd'])  #设置图表实例的属性
plt.xlabel('data set')
plt.ylabel('value')
plt.title('boxplot')
```

```python
from matplotlib.patches import Polygon

#定义求取积分的函数
def func(x):  
    return 0.5 * np.exp(x) + 1

#定义积分区间，生成必须的数值
a, b = 0.5, 1.5 # integral limits
x = np.linspace(0, 2)
y = func(x)

#绘制函数图形
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x , y, 'b' , linewidth=2)
plt.ylim(ymin=0)
                                 
# Illustrate the integral value, i.e. the area under the function
# between the lower and upper limits
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b , 0)]
poly = Polygon(verts, facecolor='0.7' , edgecolor='0.5')
ax.add_patch(poly)

#添加数学公式和一些标签
plt.text(0.5 * (a + b) , 1, r"$\int_a^b f(x)\mathrm{d}x$",
horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

#设置坐标刻度
ax.set_xticks((a, b))
ax.set_xticklabels ( ( '$a$' , '$b$'))
ax.set_yticks([func(a) , func(b)])
ax.set_yticklabels( ('$f(a)$' , '$f(b)$'))
plt.grid(True)
```
